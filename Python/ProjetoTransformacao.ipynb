# Entendendo o problema

Importe da biblioteca pandas

```python
import pandas as pd
```

Importe json dos dados de hospedagem

```python
dados = pd.read_json('/content/drive/MyDrive/Alura/dados_hospedagem.json')
```

```python
dados.head()
```

Normalização dos dados da coluna info_moveis

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

# Dados Numéricos

Separa os elementos de listas em linhas do data-frame

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[3:])
dados
```

resetando os index

```python
dados.reset_index(inplace=True, drop=True)
dados.head()
```

```python
dados.info()
```

Importe biblioteca numpy para tratar as colunas numéricas

```python
import numpy as np
```

Convertendo a coluna max_hospeder para int64

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
dados.info()
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
dados.info()
```

Removendo o caracter $

```python
dados['preco']
```

Percorre as linhas com apply
Utilizando lambda aplica o replace para substituir $ e ( , ) vírgula

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

Altera o tipo da coluna preço para float

```python
dados['preco'] = dados['preco'].astype(np.float64)
dados.info()
```

Tratando duas colunas com o apply map para substituir os caracteres

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

# Dados Textuais

```python
dados['descricao_local']
```

transforma em letras minusculas

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

removendo caracteres especiais, que não são letras de a - z ou A - Z, números 0 - 9 e hífens de palavras compostas

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

removendo os ( - ) hifens que não estão acompanhando palavras

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w).', ' ', regex=True)
```

Tokenização de string, separando as palavras em elementos de uma lista

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades']
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```
