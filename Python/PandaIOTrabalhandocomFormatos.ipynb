```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

Criando banco de dados local

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = '/content/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

Criando a tabela clientes a partir dos dados csv

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
from sqlalchemy import text
```

```python
with engine.begin() as conn:
    query = text('SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"')
    df = pd.read_sql_query(query, conn)
```

```python
df
```

```python
df.to_sql('empregados', con=engine, index=False)
```

```python
with engine.begin() as conn:
    df = pd.read_sql_table('empregados', conn)
```

```python
df
```

```python
with engine.begin() as conn:
    df = pd.read_sql_table('empregados', conn, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])

df
```

Atualizando dados no banco de dados

```python
with engine.begin() as conn:
    query = text('SELECT * FROM clientes')
    df = pd.read_sql_query(query, conn)
```

```python
with engine.begin() as conn:
    query = text('DELETE FROM clientes WHERE ID_CLIENTE = 5008804')
    conn.execute(query)
```

```python
with engine.begin() as conn:
    df = pd.read_sql_table('clientes', conn)

df
```

```python
with engine.begin() as conn:
    query = text('UPDATE clientes SET Grau_escolaridade = "Ensino superior" WHERE ID_CLIENTE = 5008808')
    conn.execute(query)
```

```python
with engine.begin() as conn:
    df = pd.read_sql_table('clientes', conn)

df
```
